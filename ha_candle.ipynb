```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
import asyncio
import plotly.graph_objects as go

```

```python
last_day =datetime.now()
last_day=last_day.date()
last_day=str(last_day)
last_day
```

```python
symbol_analyze="NVDA"


count=1200     # how much candle ?
```

```python
#################  download data


def yahoo_symbol(symbol="NVDA",timeframe='1d'):
    while True:
        try:
            data = yf.download(symbol,period='max',interval= timeframe)
            return(data)
        except Exception as e:
            print(f"An error occurred: {e}")
            asyncio.sleep(2)  # Sleep for 10 seconds before retrying




price_data=yahoo_symbol(symbol=symbol_analyze)
price_data
```

```python
price_data=price_data.tail(count)
```

```python
def heikinashi(symbol_data: pd.DataFrame) -> pd.DataFrame:
     
    df_HA = symbol_data.copy()
    df_HA['Close']=(df_HA['Open']+ df_HA['High']+ df_HA['Low']+df_HA['Close'])/4
 
    for i in range(0, len(symbol_data)):
        if i == 0:
            df_HA['Open'][i]= ( (df_HA['Open'][i] + df_HA['Close'][i] )/ 2)
        else:
            df_HA['Open'][i] = ( (df_HA['Open'][i-1] + df_HA['Close'][i-1] )/ 2)
 
    df_HA['High']=df_HA[['Open','Close','High']].max(axis=1)
    df_HA['Low']=df_HA[['Open','Close','Low']].min(axis=1)
    return df_HA
```

```python
df_HA=heikinashi(price_data)
df_HA
```

```python
df_HA['ema13']=df_HA['Close'].ewm(span=13).mean()
df_HA['sma49']=df_HA['Close'].rolling(49).mean()


df_HA
```

```python
fig = go.Figure(data=[go.Candlestick(x=df_HA.index, open=df_HA.Open, high=df_HA.High, low=df_HA.Low, close=df_HA.Close)] )
 
 
 
 
fig.add_trace(go.Scatter(x=df_HA.index, y=df_HA['ema13'], mode='lines', name='ema13', line=dict(color='black')))
fig.add_trace(go.Scatter(x=df_HA.index, y=df_HA['sma49'], mode='lines', name='sma49', line=dict(color='blue')))
 
 
 
 
 
fig.update_layout(title = "SYMBOL HA Candles", xaxis_title = 'Date', yaxis_title = 'Price',width=1200 ,height=700)
 
fig.show()
```
